```python
import matplotlib.pyplot as plt
from numpy import pi,sin,linspace,zeros,arange,cos
from ipywidgets import interact
%matplotlib inline
```

Taking a look at the Fourier series:
$$f(x) = \frac{4}{\pi} \sum^\infty_{n=odd} \frac{1}{n}\sin \left(\frac{n\pi x}{L} \right)$$
And now looking at the convergence of the series

```python
#Here we can define the square wave
def square_wave(N =1, L=5, num_points =200):
    x = linspace(-4*pi, 4*pi, num_points)
    y = zeros(num_points)
    for n in arange(1, N+1, 2):
        y = y + (1/n * sin(n*pi*x/L))
    y = y * 4*pi
    plt.plot(x,y)
square_wave(N=5, L= 5)
```

```python
#A Cool Interactive Version!
interact(square_wave,N=(1,50,2),L=5,points=200)
```
